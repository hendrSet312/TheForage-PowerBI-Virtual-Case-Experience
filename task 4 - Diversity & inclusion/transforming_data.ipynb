```python
import pandas as pd
import numpy as np

df = pd.read_excel('/content/03 Diversity-Inclusion-Dataset.xlsx', sheet_name = 'Pharma Group AG')
df.head()
```

```python
replacement = {
    "1 - Executive": 1,
    "2 - Director" : 2,
    "3 - Senior Manager" : 3,
    "4 - Manager" :  4,
    "5 - Senior Officer" : 5,
    "6 - Junior Officer": 6
}

df = df[[
    'Employee ID',
    'Job Level before FY20 promotions',
    'Job Level after FY20 promotions',
    'Job Level after FY21 promotions',
    'Gender'
]].set_index('Employee ID')



df['Job Level before FY20 promotions'] = df['Job Level before FY20 promotions'].fillna(99)
df['Job Level after FY21 promotions'] = df['Job Level after FY21 promotions'].fillna(99)

target_cols = ['Job Level before FY20 promotions','Job Level after FY20 promotions','Job Level after FY21 promotions']

for col in target_cols :
  df[col] = df[col].replace(replacement)
```

```python
df['Maximum job position reached'] = df[target_cols].min(axis = 1).replace({list(replacement.values())[i] : list(replacement.keys())[i] for i in range(6)})
```

```python
df.to_csv('maximum_pos.csv')
```
